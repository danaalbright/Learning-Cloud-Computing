```python
def increment(x):
    return x + 1

def test_increment():
    assert increment(3) == 4
    assert increment(5) == 6, "Increment failed"
    
    #Catch failures with try/except
    try:
        assert increment(-5) == -4 
    except AssertionError:
        print("Incrementing negative number failed")
        
```

```python
test_increment()
```

```python
def delivery_price(x):

    #This function computes the delivery cost given x number of miles.
    #Delivery fee within 5 miles is $5 and for over 5-10 miles is $8. 
    #We don't deliver if you are more than 10 miles away.
    price=0
    assert 0 <= x <=10, ("We only deliver within 0-10 miles!") 
    if 0<x<=5:
        price = 5
    if 5 < x <= 10:
        price=8      
            
            
    return (price)
    
def test_delivery_price():
    assert delivery_price(4) == 5
    assert delivery_price(5) == 5
    assert delivery_price(9) == 8
    try:
        assert delivery_price(13)==8
    except AssertionError:
        print("Price failed out of delivery zone")
```

```python
test_delivery_price()
```
